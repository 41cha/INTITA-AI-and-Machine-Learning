```python
import torch
import torchvision
```

```python
class EggDetection(torch.nn.Module):
    def __init__(self):
        super().__init__()
        # bs x 1 x 224 x 224
        self.conv = torch.nn.Conv2d(in_channels = 1, out_channels = 8, kernel_size = 3, padding = 1)
        # bs x 8 x 224 x 224
        self.pool = torch.nn.MaxPool2d(kernel_size = 2)
        # bs x 8 x 112 x 112
        self.linear = torch.nn.Linear(in_features = 8 * 112 * 112, out_features = 2)

    def forward(self, x:torch.Tensor):
        x = self.conv(x)
        x = torch.nn.functional.relu(x)
        x = self.pool(x)
        x = x.view(x.size(0), -1)
        x = self.linear(x)
        return x
```

```python
transform = torchvision.transforms.Compose([
    torchvision.transforms.Grayscale(1),
    torchvision.transforms.Resize((224, 224)),
    torchvision.transforms.ToTensor(),
])
```
